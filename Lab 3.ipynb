# Exercise 9
There are many variables measuring personal freedom. Consider the variables pf_ss, pf_movement, pf_religion, pf_association, pf_expression,  and pf_identity. You previously created a linear model to predict pf_score based on pf_expression. Create a multi-linear model to predict pf_score based on the variables listed. Does this provide a more accurate model than the single variable model using just pf_expression created previously?

---

Type your response here.

```python
hfi_mult = hfi_2016.filter(['pf_score', 'pf_ss', 'pf_movement', 'pf_religion', 'pf_association', 'pf_expression', 'pf_identity'])
hfi_mult_clean = hfi_mult.dropna(ignore_index=True)

mult_model_lr = LinearRegression()
X=hfi_mult_clean.drop(columns=['pf_score'])
mult_model_lr.fit(X, hfi_mult_clean['pf_score'])

for i, x in enumerate(list(X)):
    print(f"slope for {x} is {mult_model_lr.coef_[i]:.2f}")

r2 = mult_model_lr.score(X, hfi_mult_clean['pf_score'])
n = X.shape[0]
k = X.shape[1]
adjusted_r2 = 1-(1-r2)*(n-1)/(n-k-1)
print(f"The adjusted R squared value for model is {adjusted_r2}")
```

# Exercise 10
Would dropping a predictor improve the adjusted R squared for the model? Compute the adusted R squared for the models created by dropping each variable using the backward elimination technique. Which model creates the greatest adjusted R squared? What does this tell us?

---

Type your response here.

```python
for i, x in enumerate(list(X)):
    drop_mult_model_lr = LinearRegression()
    X_drop = hfi_mult_clean.drop(columns=['pf_score', x])
    drop_mult_model_lr.fit(X_drop, hfi_mult_clean['pf_score'])
    r2 = drop_mult_model_lr.score(X_drop, hfi_mult_clean['pf_score'])
    n = X_drop.shape[0]
    k = X_drop.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-k-1)
    print(f"R squared value for model that drops the variable {x} is {adjusted_r2}")
```

```python
hfi_2016['residuals'] = hfi_2016.apply(lambda row: row.pf_score - (model_lr.intercept_ + model_lr.coef_[0] * row.pf_expression_control), axis=1)
sns.histplot(data=hfi_2016['residuals'])
```
